## 1. About the Show

*Breaking Bad* is an American crime drama television series created by Vince Gilligan and broadcast on AMC. 
The series follows **Walter White** (Bryan Cranston), a high school chemistry teacher in Albuquerque, New Mexico, 
who, after being diagnosed with terminal lung cancer, turns to manufacturing and selling methamphetamine — 
together with his former student **Jesse Pinkman** (Aaron Paul) — to secure his family's financial future.

The show aired from **January 20, 2008** to **September 29, 2013**, spanning **5 seasons** and **62 episodes**. 
It is widely regarded as one of the greatest television series of all time. In 2013, *Guinness World Records* 
named it the **highest-rated TV series ever**, citing its Season 5 Metacritic score of 99 out of 100.

## 4. Viewership Over Time

```python
# Color map per season
season_colors = {1: '#e63946', 2: '#f4a261', 3: '#2a9d8f', 4: '#457b9d', 5: '#6a4c93', 6: '#1d3557'}
season_labels = {1: 'S1', 2: 'S2', 3: 'S3', 4: 'S4', 5: 'S5A', 6: 'S5B'}

fig, ax = plt.subplots(figsize=(13, 5))

for s in df['season'].unique():
    sdf = df[df['season'] == s]
    ax.plot(sdf['ep'], sdf['viewers'], color=season_colors[s], linewidth=2, label=season_labels[s])
    ax.scatter(sdf['ep'], sdf['viewers'], color=season_colors[s], s=40, zorder=5)

# Annotate series finale
ax.annotate('Series finale\n10.28M', xy=(44, 10.28), xytext=(40, 9.3),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

ax.set_title('Breaking Bad — U.S. Viewership per Episode (millions)', fontsize=14, fontweight='bold')
ax.set_xlabel('Episode number')
ax.set_ylabel('Viewers (millions)')
ax.legend(title='Season', loc='upper left')
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f M'))
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.savefig('breaking_bad_viewership.png', dpi=150)
plt.show()
```

## 5. Season-to-Season Changes in Viewership

```python
season_avg = pd.DataFrame({'season': seasons, 'avg_viewers': avg_viewers})
season_avg['change'] = season_avg['avg_viewers'].diff()
season_avg['pct_change'] = season_avg['avg_viewers'].pct_change() * 100

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: average viewers per season
colors = ['#e63946', '#f4a261', '#2a9d8f', '#457b9d', '#6a4c93', '#1d3557']
axes[0].bar(seasons, avg_viewers, color=colors, edgecolor='white', linewidth=0.5)
axes[0].set_title('Average Viewers per Season (millions)', fontweight='bold')
axes[0].set_ylabel('Viewers (millions)')
axes[0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f M'))
axes[0].grid(axis='y', linestyle='--', alpha=0.4)
for i, v in enumerate(avg_viewers):
    axes[0].text(i, v + 0.05, f'{v:.2f}M', ha='center', fontsize=9)

# Right: % change season to season
changes = season_avg['pct_change'].dropna()
change_colors = ['#2a9d8f' if x > 0 else '#e63946' for x in changes]
axes[1].bar(seasons[1:], changes, color=change_colors, edgecolor='white')
axes[1].axhline(0, color='black', linewidth=0.8)
axes[1].set_title('Season-to-Season Change in Viewership (%)', fontweight='bold')
axes[1].set_ylabel('Change (%)')
axes[1].grid(axis='y', linestyle='--', alpha=0.4)
for i, v in enumerate(changes):
    axes[1].text(i, v + 0.5, f'{v:.1f}%', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('breaking_bad_season_changes.png', dpi=150)
plt.show()
```

## 6. Analysis of Viewership Trends

*Breaking Bad* had a remarkable — and unusual — viewership trajectory. Most serialised shows lose viewers over time, but *Breaking Bad* grew continuously.

- **Season 1** averaged just **1.23 million** viewers. The pilot was partially affected by an NFL game overrun and the 2007–08 writers' strike, limiting its early reach.
- Viewership remained relatively flat through **Seasons 2 and 3**, averaging **1.54M** and **1.56M** respectively — a modest increase of only **`1.3%`** between the two seasons.
- The arrival of the show on **Netflix** before Season 4 changed everything. Season 4 averaged **2.24M** viewers, a jump of **`43.6%`** compared to Season 3.
- Season 5A continued the rise with **2.77M** average viewers (**`23.7%`** increase over Season 4).
- The most dramatic surge came with **Season 5B** (the final 8 episodes), which averaged **6.62M** viewers — a staggering **`139%`** increase over Season 5A.
- The **series finale** (*Felina*) drew **10.28 million** viewers, making it the most-watched episode in the show's history and one of the most-watched cable drama finales ever.

Overall, from Season 1 to Season 5B, average viewership increased by **`438%`** — a growth story almost unparalleled in television history.
