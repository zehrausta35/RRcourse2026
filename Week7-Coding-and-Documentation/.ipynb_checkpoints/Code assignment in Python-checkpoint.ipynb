```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.ticker as ticker
from matplotlib.gridspec import GridSpec
```

```python
DATA_FOLDER = "../Data"
ONET_FILE = f"{DATA_FOLDER}/onet_tasks.csv"
EUROSTAT_FILE = f"{DATA_FOLDER}/Eurostat_employment_isco.xlsx"

# Countries to analyse
COUNTRIES = ["Belgium", "Spain", "Poland"]

# ISCO occupation group numbers (1 through 9)
ISCO_GROUPS = list(range(1, 10))

# Task variables for Non-Routine Cognitive Analytical (NRCA) tasks
NRCA_TASKS = ["t_4A2a4", "t_4A2b2", "t_4A4a1"]

task_data = pd.read_csv(ONET_FILE)

# Each sheet has employment by quarter for one ISCO occupation group
isco_sheets = []
for group_num in ISCO_GROUPS:
    sheet = pd.read_excel(EUROSTAT_FILE, sheet_name=f"ISCO{group_num}")
    sheet["ISCO"] = group_num  # label which occupation group this row belongs to
    isco_sheets.append(sheet)

# Combine all occupation groups into one table
all_data = pd.concat(isco_sheets, ignore_index=True)

# calculate each country's total workers per time period
for country in COUNTRIES:
    # Sum workers across all occupation groups for this country
    total_per_group = [sheet[country] for sheet in isco_sheets]
    country_total = sum(total_per_group)

    # Repeat the total 9 times (once per occupation group) so it lines up with all_data
    all_data[f"total_{country}"] = pd.concat([country_total] * len(ISCO_GROUPS), ignore_index=True)

    # Share = workers in this occupation / all workers in this country
    all_data[f"share_{country}"] = all_data[country] / all_data[f"total_{country}"]

# Keep only the first digit of the ISCO code so it matches our 1-digit groups
task_data["isco08_1dig"] = task_data["isco08"].astype(str).str[:1].astype(int)

# Average task values within each 1-digit occupation group
aggdata = task_data.groupby("isco08_1dig").mean().drop(columns=["isco08"])


# merge employment and task data
combined = pd.merge(all_data, aggdata, left_on="ISCO", right_on="isco08_1dig", how="left")

# Standardising = shifting so mean is 0 and spread (std dev) is 1
# We use employment shares as weights so larger occupations count more
for country in COUNTRIES:
    weight_col = f"share_{country}"
    for task in NRCA_TASKS:
        # Weighted mean
        w_mean = np.average(combined[task], weights=combined[weight_col])
        # Weighted standard deviation
        w_sd = np.sqrt(np.average((combined[task] - w_mean) ** 2, weights=combined[weight_col]))
        # Store the standardised value
        combined[f"std_{country}_{task}"] = (combined[task] - w_mean) / w_sd

for country in COUNTRIES:
    std_task_cols = [f"std_{country}_{task}" for task in NRCA_TASKS]
    combined[f"{country}_NRCA"] = combined[std_task_cols].sum(axis=1)
```

```python
# standardise the NRCA score itself
for country in COUNTRIES:
    weight_col = f"share_{country}"
    nrca_col = f"{country}_NRCA"

    w_mean = np.average(combined[nrca_col], weights=combined[weight_col])
    w_sd = np.sqrt(np.average((combined[nrca_col] - w_mean) ** 2, weights=combined[weight_col]))
    combined[f"std_{country}_NRCA"] = (combined[nrca_col] - w_mean) / w_sd

# This gives a single NRCA value per country per quarter
agg_results = {}
for country in COUNTRIES:
    # Weighted value = standardised NRCA * share of workers in that occupation
    combined[f"multip_{country}_NRCA"] = combined[f"std_{country}_NRCA"] * combined[f"share_{country}"]

    # Sum across occupations for each time period
    agg_results[country] = combined.groupby("TIME")[f"multip_{country}_NRCA"].sum().reset_index()
```

```python
# plot results
for country in COUNTRIES:
    agg = agg_results[country]
    time_col = "TIME"
    value_col = f"multip_{country}_NRCA"
    plt.figure()
    plt.plot(agg[time_col], agg[value_col])
    plt.title(f"Non-Routine Cognitive Analytical Task Intensity – {country}")
    plt.xlabel("Time")
    plt.ylabel("NRCA Index")
    plt.xticks(range(0, len(agg), 3), agg[time_col][::3], rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Color palette - one distinct color per country
COUNTRY_COLORS = {
    "Belgium": "#E63946",
    "Spain":   "#2A9D8F",
    "Poland":  "#E9C46A",
}

fig = plt.figure(figsize=(16, 10))
fig.patch.set_facecolor("#0F1117")

gs = GridSpec(2, 3, figure=fig, hspace=0.5, wspace=0.35)

axes_individual = [fig.add_subplot(gs[0, i]) for i in range(3)]
ax_combined     = fig.add_subplot(gs[1, :])   # spans full bottom row

# individual country plots
for ax, country in zip(axes_individual, COUNTRIES):
    agg        = agg_results[country]
    time_col   = "TIME"
    value_col  = f"multip_{country}_NRCA"
    color      = COUNTRY_COLORS[country]

    # Shaded area under the line
    ax.fill_between(agg[time_col], agg[value_col], alpha=0.15, color=color)
    ax.plot(agg[time_col], agg[value_col], color=color, linewidth=2)

    # Highlight the max point
    max_idx = agg[value_col].idxmax()
    ax.scatter(agg.loc[max_idx, time_col], agg.loc[max_idx, value_col],
               color=color, s=60, zorder=5)
    ax.annotate(f"peak: {agg.loc[max_idx, value_col]:.2f}",
                xy=(agg.loc[max_idx, time_col], agg.loc[max_idx, value_col]),
                xytext=(8, 6), textcoords="offset points",
                fontsize=7.5, color=color)

    # Styling
    ax.set_facecolor("#1A1D27")
    ax.set_title(country, fontsize=13, fontweight="bold", color="white", pad=10)
    ax.set_xlabel("Quarter", fontsize=8, color="#AAAAAA")
    ax.set_ylabel("NRCA Index", fontsize=8, color="#AAAAAA")
    ax.tick_params(colors="#AAAAAA", labelsize=7)
    ax.set_xticks(range(0, len(agg), 4))
    ax.set_xticklabels(agg[time_col][::4], rotation=45, ha="right")
    ax.yaxis.set_major_formatter(ticker.FormatStrFormatter("%.2f"))
    for spine in ax.spines.values():
        spine.set_edgecolor("#333344")
    ax.grid(axis="y", color="#2A2D3A", linewidth=0.7, linestyle="--")
    ax.grid(axis="x", color="#2A2D3A", linewidth=0.4, linestyle=":")

ax_combined.set_facecolor("#1A1D27")

for country in COUNTRIES:
    agg       = agg_results[country]
    time_col  = "TIME"
    value_col = f"multip_{country}_NRCA"
    color     = COUNTRY_COLORS[country]

    ax_combined.fill_between(agg[time_col], agg[value_col], alpha=0.08, color=color)
    ax_combined.plot(agg[time_col], agg[value_col],
                     label=country, color=color, linewidth=2)

ax_combined.set_title("All Countries — NRCA Comparison", fontsize=13,
                       fontweight="bold", color="white", pad=10)
ax_combined.set_xlabel("Quarter", fontsize=9, color="#AAAAAA")
ax_combined.set_ylabel("NRCA Index", fontsize=9, color="#AAAAAA")
ax_combined.tick_params(colors="#AAAAAA", labelsize=7.5)
ax_combined.set_xticks(range(0, len(agg), 4))
ax_combined.set_xticklabels(agg[time_col][::4], rotation=45, ha="right")
ax_combined.yaxis.set_major_formatter(ticker.FormatStrFormatter("%.2f"))
for spine in ax_combined.spines.values():
    spine.set_edgecolor("#333344")
ax_combined.grid(axis="y", color="#2A2D3A", linewidth=0.7, linestyle="--")
ax_combined.grid(axis="x", color="#2A2D3A", linewidth=0.4, linestyle=":")
ax_combined.legend(fontsize=10, facecolor="#1A1D27",
                   edgecolor="#333344", labelcolor="white")

fig.suptitle("Non-Routine Cognitive Analytical (NRCA) Task Intensity",
             fontsize=16, fontweight="bold", color="white", y=1.01)

plt.savefig("nrca_dashboard.png", dpi=150, bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.show()
```
